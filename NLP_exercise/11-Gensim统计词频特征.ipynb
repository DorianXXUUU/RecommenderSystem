```python
from universalMethod import *
import gensim
from gensim import corpora
```

```python
def gensimCorpus(corpus=None):
    dictionary = corpora.Dictionary(corpus)
    dfs = dictionary.dfs
    for key_id, freq in dfs.items():
        print(dictionary[key_id], '\t', freq)
```
