```python
import os
from universalMethod import *
from gensim import corpora, models
```

```python
def gensimCorpus(corpus=None, classVec=''):
    dicionary = corpora.Dictionary(corpus)
    # 生成词袋模型
    docBagCorpus = [dicionary.doc2bow(docCut) for docCut in corpus]
    print("词袋模型实现文本向量化\n", docBagCorpus)
    # 生成tfidf
    tfidfModel = models.TfidfModel(dictionary=dicionary)
    print(tfidfModel)
    
    # tfidf字典
    corpus_tfidf = {}
    i = 0 # 获取类别
    for doc_bow in docBagCorpus:
        # 词袋填充
        file_tdidf = tfidfModel[doc_bow]
        catg = classVec[i]
        tmp = corpus_tfidf.get(catg, [])
        tmp.append(file_tdidf)
        print('Class NO.', i, ' feature words tf-idf\n', tmp)
        if tmp.__len__() == 1:
            corpus_tfidf[catg] = tmp
        i+=1
    # 本地化储存
    print('TF-IDF: \n', corpus_tfidf)
    catgs = list(corpus_tfidf.keys())
    for catg in catgs:
        savePath = r'../dataSet/files/tfidf_corpus'
        corpora.MmCorpus.serialize('{f}{s}{c}.mm'.format(f = savePath,
                                                         s = os.sep, 
                                                         c = catg), corpus_tfidf.get(catg), id2word=dicionary)
```

```python
if __name__ == '__main__':
    corpus, classVec = loadDataSet()
    gensimCorpus(corpus, classVec)

#     TfidfModel(num_docs=5, num_nnz=568) -- (总共五类文档, 整个文件不重复词的个数)
```
